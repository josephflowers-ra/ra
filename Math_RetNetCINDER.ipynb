```python
%cd /content/RetNet

import torch
from retnet.modeling_retnet import RetNetModelWithLMHead
from retnet.configuration_retnet import load_config_from_yaml
from transformers import AutoTokenizer

# Uncomment to train model from scratch:
#config = load_config_from_yaml('/content/RetNet/configs/retnet-mini.yml')
#model = RetNetModelWithLMHead(config)

# Uncomment to continue training from the pretrained model:
model = RetNetModelWithLMHead.from_pretrained('/content/RetNetCINDER/checkpoint-2500')

model_size = sum(t.numel() for t in model.parameters())
print(f"Model size: {model_size/1000**2:.1f}M parameters")

tokenizer = AutoTokenizer.from_pretrained("tiiuae/falcon-7b")
tokenizer.add_special_tokens({'pad_token': '<|endoftext|>'})
tokenizer.pad_token = tokenizer.eos_token
```

```python
%cd /content/RetNet

# Run this cell for training

from datasets import load_dataset, DatasetDict

dataset = load_dataset(
  "text",
  data_files={ "train": [ "/content/drive/MyDrive/simplemathh.txt" ], "test": [ "/content/drive/MyDrive/mathtest.txt" ] },
  sample_by="paragraph")

raw_datasets = DatasetDict(
    {
        "train": dataset["train"],  # .shuffle().select(range(50000)),
        "valid": dataset["test"],  # .shuffle().select(range(500))
    }
)

from transformers import Trainer, TrainingArguments
from transformers import DataCollatorForLanguageModeling

data_collator = DataCollatorForLanguageModeling(tokenizer, mlm=False)

def tokenize(element):
    outputs = tokenizer(
        element["text"],
        truncation=True,
        max_length=256,
        return_overflowing_tokens=True,
        return_length=True)

    input_batch = []
    for length, input_ids in zip(outputs["length"], outputs["input_ids"]):
        input_batch.append(input_ids)

    return {"input_ids": input_batch}

tokenized_datasets = raw_datasets.map(
    tokenize, batched=True, remove_columns=raw_datasets["train"].column_names)

args = TrainingArguments(
    output_dir="/content/RetNetCINDER",
    per_device_train_batch_size=32,
    per_device_eval_batch_size=32,
    evaluation_strategy="steps",
    eval_steps=500,
    logging_steps=500,
    gradient_accumulation_steps=8,
    num_train_epochs=5,
    weight_decay=0.01,
    warmup_steps=500,
    lr_scheduler_type="cosine",
    learning_rate=6e-4,
# checkpointing:
    save_steps=1000,
    optim="adamw_torch",
    fp16=True,
#    push_to_hub=True,
)

trainer = Trainer(
    model=model,
    args=args,
    tokenizer=tokenizer,
    data_collator=data_collator,
    train_dataset=tokenized_datasets["train"],
    eval_dataset=tokenized_datasets["valid"],
)

trainer.train()
trainer.save_model()
```

```python
%cd /content/RetNet

# Run this cell for inference (text generation)

import torch
from torch import nn

from transformers import AutoTokenizer
from transformers import top_k_top_p_filtering
from retnet.modeling_retnet import RetNetModelWithLMHead

tokenizer = AutoTokenizer.from_pretrained("tiiuae/falcon-7b")
tokenizer.add_special_tokens({'pad_token': '<|endoftext|>'})
tokenizer.pad_token = tokenizer.eos_token

# Comment out the following line to reuse the model trained in cell above
model = RetNetModelWithLMHead.from_pretrained('/content/RetNetCINDER/checkpoint-13500')
#model = model.to(device='cuda:0')

inputs = tokenizer("multiply 5 and 10.\n", return_tensors='pt', return_token_type_ids=False)
input_ids = inputs.input_ids
#input_ids = input_ids.to(device='cuda:0')

import time
print(tokenizer.decode(input_ids[0]), end="")

torch.manual_seed(42)
start_time = time.time()

max_new_tokens = 128

# We can simply process the prompt (in parallel) and recurrently generate new tokens with:
output = model.generate(**inputs, max_new_tokens=max_new_tokens, do_sample=True, top_p=0.99)
print(tokenizer.decode(output[0]))
end_time = time.time()
print("time cost:", int(float(end_time - start_time) * 1000.0), "ms", int(float(end_time - start_time) * 1000.0 / max_new_tokens), "ms/tok")
```

```python
%cd /content/RetNet

# Same as the above model.generate showing some details from under the hood:
torch.manual_seed(42)
start_time = time.time()

inputs = tokenizer("CINDER: ", return_tensors='pt', return_token_type_ids=False)
input_ids = inputs.input_ids

# Process initial prompt

past_key_values = None
for i in range(input_ids.shape[1]):
#    print(tokenizer.decode(input_ids[:, i:i+1][0]), end='')
    rnn_out = model(input_ids[:, i:i+1], forward_impl='recurrent', past_key_values=past_key_values, use_cache=True, sequence_offset=i)#, output_hidden_states=True)
    past_key_values = rnn_out.past_key_values

# Recursively generate more tokens

for j in range(input_ids.shape[1], input_ids.shape[1] + max_new_tokens):
    logits = rnn_out.logits[:,-1]
    next_token = model.sample_token(logits, do_sample=True, top_k=0, top_p=0.99)
#    print(tokenizer.decode(next_token[0]), end='')
    input_ids = torch.cat([input_ids, next_token], dim=-1)
    rnn_out = model(input_ids[:, j:j+1], forward_impl='recurrent', past_key_values=past_key_values, use_cache=True, sequence_offset=j)
    past_key_values = rnn_out.past_key_values

end_time = time.time()
print(tokenizer.decode(input_ids[0]))
print("time cost:", int(float(end_time - start_time) * 1000.0), "ms", int(float(end_time - start_time) * 1000.0 / max_new_tokens), "ms/tok")
```
